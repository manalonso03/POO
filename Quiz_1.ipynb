```python
# Pregunta 3
import pandas as pd
df= pd.DataFrame(diccionario)
df
# Atraves del data frame se puede conseguir una tabla del diccionario antes hecho.
```

```python
#primera segregación
df.drop(columns=["Alejandra"])
#Con la función drop se quitan las columnas o labels que no querramos
```

```python
#segunda segregación
df.drop(index=1)
```

```python
# tercera segregación
df.drop(labels=4)
```

```python
# Pregunta 4
df.describe()
#Con la función describe se nos describe como se comportan estos datos, por ejemplo da los calores maximos y minimos de las tres llaves.
```
