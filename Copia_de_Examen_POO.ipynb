**Ejercicio 4:** A partir de un diccionario, crea un DataFrame (que llamarás **precios_muestra**) que consista de dos columnas y que tenga 12 valores **flotantes** que eligirás de manera aleatoria en el intervalo (8,13). Etiqueta del 1 al 12 y llama a la columna de etiquetas Mes.

1.- Construye una función que calcule los rendimientos simples de cada columna.

2.- Las desviaciones estadísticas de los rendimientos ayudan a calcular la volatilidad del portafolio:
$$\text{desviaciones}=(\text{rendimientos}-\text{media}(\text{rendimientos}))^2.$$
Crea una función que calcule la media de las desviaciones y luego la volatilidad de tu portafolio (raíz cuadrada de la media de las desviaciones).

3.- Anualizamos la volatilidad escalándola (multiplicándola) por la raíz cuadrada del número de períodos por observación.

Por lo tanto, para anualizar la volatilidad de una serie mensual, la multiplicamos por la raíz cuadrada de 12. Muestra la serie de la volatilidad anualizada.

4.- Realiza una gráfica de barras para los rendimientos y una gráfica en líneas para los precios.

5.- Calcula los rendimientos compuestos para cada columna y después anualiza los rendimientos.

---

```python
# Importamos lo necesario
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creamos el DataFrame con datos aleatorios.

data = {'Mes': range(1, 13),
        'Precio': np.random.uniform(8, 13, 12)}
precios_muestra = pd.DataFrame(data) # Le damos nombre a nuestro Data Frame.

# Creamos la Función para calcular rendimientos simples.
def calculo_rendimientos(dataframe): #Definimos nuestra función
    dataframe['Rendimiento'] = dataframe['Precio'].pct_change() #Con la función multiplicamos el precio por el cambio que tuvo la acción de cada fila.
    return dataframe

precios_muestra = calculo_rendimientos(precios_muestra)

# Creamos la Función para calcular la volatilidad anualizada
def calculo_volatilidad(dataframe): #Definimos nuestra función
    desviacion = (dataframe['Rendimiento'] - dataframe['Rendimiento'].mean())**2 # La función de la desviación resta el rendimiento del rendimiento por la media y lo elevamos al cuadrado.
    media_desviaciones = desviacion.mean() # Sacamos la media de las desviaciones con la función mean() para poder anualizarla.
    volatilidad = np.sqrt(media_desviaciones) * np.sqrt(12) # Anualizaqmos la volatilidad con la raiz cuadrada del numero de periodos.
    return volatilidad

volatilidad_anualizada = calculo_volatilidad(precios_muestra)

# Creamos las gráficas de rendimientos y precios, através de las funciones plt, por eso importamos el matplot.lib.pyplot.
plt.figure(figsize=(12, 5)) #Definimos el tamaño.
plt.subplot(1, 2, 1)
plt.bar(precios_muestra['Mes'], precios_muestra['Rendimiento'])
plt.xlabel('Mes') # Damos los ejes Y y X.
plt.ylabel('Rendimiento')

plt.subplot(1, 2, 2)
plt.plot(precios_muestra['Mes'], precios_muestra['Precio'])
plt.xlabel('Mes')
plt.ylabel('Precio')

plt.tight_layout()
plt.show()

# Damos la función para calcular rendimientos compuestos y anualizarlos
def calcular_rendimientos_anualizados(dataframe): #Definimos la función.
    rendimientos_compuestos = (dataframe['Rendimiento'] + 1).cumprod() - 1
    rendimientos_mensuales = dataframe['Rendimiento'] + 1 # Sacamos los rendimientos mensuales para póder anualizarlos
    rendimientos_promedio = rendimientos_mensuales.prod() ** (12 / len(dataframe)) - 1 #Sacamos su promedio
    rendimientos_anualizados = rendimientos_promedio * 12 #Los anualizamos
    return rendimientos_anualizados

rendimientos_anualizados = calcular_rendimientos_anualizados(precios_muestra)

# Resultados
print("Volatilidad Anualizada:", volatilidad_anualizada)
print("Rendimientos Anualizados:", rendimientos_anualizados)
```
